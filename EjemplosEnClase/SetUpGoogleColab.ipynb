```python
import os
import sys
```

```python
pathendrive = '/content/drive/Shareddrives/Modelizacion/'
repository = 'https://github.com/Modelizacion-de-Materiales/2022_Ingenieria.git'
if os.path.exists(pathendrive):
  os.chdir(pathendrive)
else:
  raise FileNotFoundError('La carpeta indicada no existe')

if not os.path.exists(os.path.join(pathendrive, '2022_Ingenieria')):
  os.system('git clone '+repository)
else:
  raise Warning('Ya tiene el repositorio descargado')
```
